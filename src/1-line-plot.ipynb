### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]  # Squaring the x values

# Create the line plot
plt.plot(x, y, marker='o', color='blue', linestyle='-', linewidth=2)

# Customize the plot
plt.title("Line Plot: Squares of X")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
